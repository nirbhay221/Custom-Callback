```
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt 
import datetime
import io
import imageio
from PIL import Image
from IPython.display import Image as IPyImage
```

```
def get_model():
  model = tf.keras.models.Sequential()
  model.add(tf.keras.layers.Dense(1,activation='linear',input_dim=784))
  model.compile(optimizer= tf.keras.optimizers.RMSprop(lr=0.1),loss= 'mean_squared_error',metrics=['mae'])
  return model
```

```
(x_train,y_train),(x_test,y_test)= tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(60000,784).astype('float32')/255.0
x_test = x_test.reshape(10000,784).astype('float32')/255.0
```

```
class MyCallback(tf.keras.callbacks.Callback):
  def on_train_batch_begin(self,batch,logs=None):
    print('Training: batch {} begins at {}'.format(batch,datetime.datetime.now().time()))
  def on_train_batch_end(self,batch,logs=None):
    print('Testing : batch {} ends at {}'.format(batch,datetime.datetime.now().time()))
```

```
model = get_model()
_ = model.fit(x_train,y_train,batch_size=64,epochs=1,steps_per_epoch=5,verbose=0,callbacks=[MyCallback()])
```

```
callback = tf.keras.callbacks.LambdaCallback(on_epoch_end= lambda epoch,logs:print('Epoch: {} ,Val/Train loss ratio :{:.2f}'.format(epoch,logs['val_loss']/logs['loss'])))
```

```
model = get_model()
_= model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs= 3,verbose=0,batch_size=64,callbacks=[callback])
```

```
class DetectingOverfittingCallback(tf.keras.callbacks.Callback):
  def __init__(self,threshold=0.7):
    super(DetectingOverfittingCallback,self).__init__()
    self.threshold=threshold
  def on_epoch_end(self,epochs,logs=None):
    ratio = logs['val_loss']/logs['loss']
    print("Epoch: {}, Val/Train loss ratio: {:.2f}".format(epochs, ratio))
    if ratio>self.threshold:
      print('stop training')
      self.model.stop_training = True
model= get_model()
_= model.fit(x_train,y_train,validation_data= (x_test,y_test),epochs=3,batch_size=64,callbacks=[DetectingOverfittingCallback()],verbose=0)
```

```
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(60000,784).astype('float32')/255.0
x_test = x_test.reshape(10000,784).astype('float32')/255.0
```

```
plt.rc('font',size=20)
plt.rc('figure',figsize=(15,3))
def display_digits(inputs,outputs,ground_truth,epoch,n=10):
  plt.clf()
  plt.yticks([])
  plt.grid(None)
  inputs = np.reshape(inputs,[n,28,28])
  inputs = np.swapaxes(inputs,0,1)
  inputs = np.reshape(inputs,[28,28*n])
  plt.imshow(inputs)
  plt.xticks([28*x+14 for x in range(n)],outputs)
  for i,t in enumerate(plt.gca().xaxis.get_ticklabels()):
    if outputs[i]==ground_truth[i]:
      t.set_color('green')
    else:
      t.set_color('red')
  plt.grid(None)
```

```
GIF_PATH = './animation.gif'
```

```
class VisCallbacks(tf.keras.callbacks.Callback):
  def __init__(self,inputs,ground_truth,display_freq=10,n_samples=10):
    self.inputs =inputs
    self.ground_truth = ground_truth
    self.display_freq= display_freq
    self.images=[]
    self.n_samples= n_samples
  def on_epoch_end(self,epochs,logs=None):
    indexes = np.random.choice(len(self.inputs),size=self.n_samples)
    x_test,y_test =self.inputs[indexes],self.ground_truth[indexes]
    predictions = np.argmax(self.model.predict(x_test),axis=1)
    display_digits(x_test,predictions,y_test,epochs,n=self.display_freq)
    buf = io.BytesIO()
    plt.savefig(buf,format='png')
    buf.seek(0)
    image = Image.open(buf)
    self.images.append(np.array(image))
    if epochs%self.display_freq ==0:
      plt.show()
  def on_train_end(self,logs=None):
    imageio.mimsave(GIF_PATH,self.images,fps=1)


```

```
def get_model():
  model = tf.keras.models.Sequential()
  model.add(tf.keras.layers.Dense(32,activation='linear',input_dim=784))
  model.add(tf.keras.layers.Dense(10,activation='softmax'))
  model.compile(optimizer= tf.keras.optimizers.RMSprop(lr=1e-4),loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
  return model
```

```
model = get_model()
_=model.fit(x_train,y_train,epochs=20,batch_size=64,verbose=0,callbacks=[VisCallbacks(x_test,y_test)])
```
